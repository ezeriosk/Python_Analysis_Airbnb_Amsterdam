## DATA UPLOAD

```python
calendar = pd.read_csv('Datasets/calendar.csv')
listings_details = pd.read_csv('Datasets/listings_details.csv')
listings = pd.read_csv('Datasets/listings.csv');
```

## DATA PREPARATION

### Calendar

```python
calendar.head(2)
```

```python
calendar['year'] = pd.to_datetime(calendar['date']).dt.year
calendar['month'] = pd.to_datetime(calendar['date']).dt.month_name()
```

```python
calendar2=pd.DataFrame(calendar.groupby(['listing_id','year','month','available']).count()['date']).reset_index()
```

```python
calendar2.rename(columns={'date':'total','listing_id':'id'},inplace=True)
calendar2['id'] = calendar2['id'].astype('str')
```

```python
calendar2.to_csv('calendar2.csv',index=False)
```

### Listings + Listings details

```python
listings2 = listings[['id','name','neighbourhood','latitude','longitude','room_type','price',
                      'number_of_reviews','reviews_per_month','availability_365']]
```

```python
#List of columns that i want to use

listings_details_to_use = ['id','host_is_superhost','neighbourhood',
                'latitude','longitude', 'property_type','room_type','bedrooms',
                 'price','security_deposit','cleaning_fee', 
                  'number_of_reviews','reviews_per_month']

not_in_listings_details_columns = []

# List of columns that are not repeated 

for column in listings_details.columns:
    if column not in listings2.columns:
        not_in_listings_details_columns.append(column)
        
listings_details_to_use2 = []   

# Columns that i want to use and are not repeated with listings

for column in listings_details_to_use:
    if column in not_in_listings_details_columns:
        listings_details_to_use2.append(column)
        
listings_details_to_use2.append('id')

# Leave only the columns that i want

listings_details2 = listings_details[listings_details_to_use2]
```

```python
listings_complete = listings2.merge(listings_details2,on='id')
```

```python
status(listings_complete)
```

```python
listings_complete[['id','latitude','longitude']]=listings_complete[['id','latitude','longitude']].astype('str')
```

```python
listings_complete['security_deposit'] = listings_complete['security_deposit'].str[1:-3].fillna(0).str.replace(',','').astype('float')
listings_complete['cleaning_fee'] = listings_complete['cleaning_fee'].str[1:-3].fillna(0).str.replace(',','').astype('float')
```

```python
listings_complete.to_csv('listings_complete.csv',index=False)
```
